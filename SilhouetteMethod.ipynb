```python
from sklearn.metrics import silhouette_score
from yellowbrick.cluster import SilhouetteVisualizer
import numpy as np
from sklearn.cluster import KMeans
import pandas as pd
from sklearn import metrics
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist



xaxis = np.array([1.8, 2.8, 5, 5.6, 1.7, 4.5, 5, 5.4, 1.5, 2, 4.8, 5.5, 5, 2.7, 2.2, 1.7, 2.2, 4.2, 2, 5.6, 4, 1.7, 5])
yaxis = np.array([5, 4, 4, 3.5, 4, 8.5, 5, 4, 4, 4, 8, 5, 3.5, 3.8, 7.4, 5, 3.5, 8, 7, 4.5, 9, 6.8, 4.4])
X = np.array(list(zip(xaxis, yaxis))).reshape(len(xaxis), 2)
plt.plot()
plt.title('Data Visualization')
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(xaxis, yaxis)
plt.show()
#plt.savefig('output/histogram.png')

#defined number of clusters here for Kmeans
kmeanModel = KMeans(n_clusters=4)

#Silhouette plot, fit the data to the silhouette visualizer
silhouette = SilhouetteVisualizer(kmeanModel,colors = 'yellowbrick')
plt.title("Silhouette Plot")
plt.gca().axes.set_yticklabels([])
silhouette.fit(X)  
plt.xlabel('Silhouette Coefficient')
plt.ylabel('Cluster Levels')   
plt.show()

#fit the Kmean model
kmeanModel.fit(X)
label= kmeanModel.predict(X)

#silhouette coefficient
Silhouette__Score = silhouette_score(X, label)
print ("Silhouette Score is: " +  str(Silhouette__Score))

#To display clustered scatter plot for better understanding
unique_labels = np.unique(label)
plt.title('Data Visualization with Clusters')
for c in unique_labels:
    plt.scatter(X[label == c , 0] , X[label == c , 1] , label = c)
#plt.savefig('output/h.png')
plt.show()
 
```
